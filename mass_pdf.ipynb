```python
import pandas as pd
from datetime import date
import numpy as np
from PIL import ImageFont, ImageDraw, Image
```

```python
df=pd.read_csv("names.csv")
```

```python
font = ImageFont.truetype("OpenSans-Regular.ttf", 70)
date_font=ImageFont.truetype("OpenSans-Regular.ttf", 40)
```

```python
df.head()
```

```python
for i in range(0, len(df)):
    img = Image.open('certificate.jpg')
    I1 = ImageDraw.Draw(img)
    I1.text((1500, 1300), df.loc[i].Name, font=font, fill =(0, 0, 0))
    img.save(df.loc[i].Name+ " certificate.png")
```

```python
for i in range(0, len(df)):
    img=Image.open(df.loc[i].name+" certificate.png")
    im1=img.convert("RGB")
    im1.save(df.loc[i].name+" certificate.pdf")
```
